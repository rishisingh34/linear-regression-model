```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.axes as ax  
df = pd.read_csv('data_for_lr.csv')
df.head(10)
```

```python
df.info()
```

```python
print("Original Shape :", df.shape) 
df = df.dropna()
print("Shape after dropping rows with missing values :", df.shape) 
```

```python
test_input = np.array(df.x[500:700]).reshape(199, 1)
test_output = np.array(df.y[500:699]).reshape(199, 1)
train_output = np.array(df.y[0:500]).reshape(500, 1)
train_input = np.array(df.x[0:500]).reshape(500, 1)
```

```python
print(f'Shape of train Input : {train_input.shape} and Shape of train_output : {train_output.shape}')
print(f'Shape of test_input : {test_input.shape} and Shape of test_output : {test_output.shape}')
```

<h1 style='color : gray ; font-weight : 700 ;'>Linear Regression - Scikit Learn </h1>

<h2 style ='color : dimgray ; '>Training Model</h2>

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr_model = LinearRegression() 
```

```python
lr_model.fit(train_input, train_output)
```

```python
lr_model.coef_ # giving the 'm' parameter : the slope 
```

<h2 style ='color : dimgray ; '>Prediction</h2>

```python
test_predictions = lr_model.predict(test_input)
```

```python
plt.plot(test_input, test_predictions, '+', color = 'blue')
plt.plot(test_input, test_output, '*', color = 'orange')
plt.title('Prediction Analysis')
plt.xlabel('Input')
plt.ylabel('Output/Predictions')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
cost = mean_squared_error(test_output, test_predictions)
cost 
```
